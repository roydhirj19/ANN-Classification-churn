```python
import tensorflow as tf
import joblib
import numpy as np
import pandas as pd
```

```python
## load the train model.scaler ,pickle,onehot
model = tf.keras.models.load_model('ann_model.keras')

scaler = joblib.load('scaler.pkl')
gender_encoder = joblib.load('gender_encoder.pkl')
geo_encoder = joblib.load('geo_encoder.pkl')
```

```python
## input data
input_data = {
    "CreditScore" : 600,
    "Age" : 40,
    "Tenure" : 3,
    "Balance" : 60000 ,
    "Geography" :"France",
    "Gender" : "Male" ,    
    "NumOfProducts" : 2,
    "HasCrCard" : 1,
    "IsActiveMember" : 1,
    "EstimatedSalary": 50000
}
```

```python
# print(scaler.feature_names_in_)
# print(input_df.columns)
```

```python
input_df = pd.DataFrame([input_data])
```

```python
## convert data into one hot encoding
gender_encoded = gender_encoder.transform(input_df[['Gender']])
gender_encoder_df = pd.DataFrame(gender_encoded,columns=gender_encoder.get_feature_names_out())
```

```python
geo_encoded = geo_encoder.transform(input_df[['Geography']])
geo_encoder_df = pd.DataFrame(geo_encoded,columns=geo_encoder.get_feature_names_out())
```

```python
## combine one hot with input data
input_df = pd.concat([input_df,gender_encoder_df,geo_encoder_df],axis=1)
```

```python
input_df = input_df.drop(['Geography','Gender'],axis=1)
```

```python
input_scaled = scaler.transform(input_df)
```

```python
prediction = model.predict(input_scaled)
```

```python
print(prediction)
```

```python
prediction_probo = prediction[0][0]
```
